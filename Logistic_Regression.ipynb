```python
from sklearn.linear_model import LogisticRegression
# it imports ml model -Logistic Regression which is used for classification tasks
# Logistic Regression is used when the output is binary (0 or 1). It gives us the probability that something belongs to a certain class

import numpy as np
# numpy to create and handle numeric data in arrays

X = np.array([[1],[2],[3],[4],[5],[6],[7]])
y= np.array([0,0,0,1,1,1,1])

model = LogisticRegression()
model.fit(X,y)

hours = np.array([2]).reshape(-1, 1) # Reshape hours to be a 2D array

prediction = model.predict(hours)

print(f"Prediction for {hours[0][0]} hours of study:", "Pass" if prediction[0] else "Fail")
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

X_test = np.linspace(0, 8, 100).reshape(-1, 1)
y_pred = model.predict_proba(X_test)[:, 1]

plt.scatter(X, y, label='Data Points')
plt.plot(X_test, y_pred, color='red', label='Logistic Regression')
plt.xlabel('Hours of Study')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5]
y=[10,20,25,35,50]

plt.plot(x,y)
plt.title("Study Hours vs score")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.show()
```
